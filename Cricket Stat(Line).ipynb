```python
# aggregation for Runs scored

player_stats = df.groupby('Player').agg({'Match': 'sum',
                                         'Innings': 'sum',
                                         'Not Outs': 'sum',
                                         'Runs': 'sum',
                                         'Balls Faced': 'sum',
                                         '100s': 'sum',
                                         '50s': 'sum',
                                         'Ducks': 'sum',
                                         '4s': 'sum',
                                         '6s': 'sum',
                                         'Average': 'sum',
                                         'SR' : 'max',
                                         'High Score': 'sum',
                                         'Innings by Bowling': 'sum',
                                         'Balls Bowled' : 'sum',
                                         'Maiden' : 'sum',
                                         'Runs Given' : 'sum',
                                         'Wkts' : 'sum',
                                         '4 W ' : 'sum',
                                         '5 W' : 'sum',
                                         'Catches' : 'sum',
                                         'Stimping' : 'sum',
                                         'B Averag' : 'sum',
                                         'B SR' : 'max',
                                         'Econ' : 'min'}).reset_index()

player_stats.head()
```
