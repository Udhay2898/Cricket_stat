```python
# aggregation for Runs scored

player_stats = df.groupby('Player').agg({'Match': 'sum',
                                         'Innings': 'sum',
                                         'Not Outs': 'sum',
                                         'Runs': 'sum',
                                         'Balls Faced': 'sum',
                                         '100s': 'sum',
                                         '50s': 'sum',
                                         'Ducks': 'sum',
                                         '4s': 'sum',
                                         '6s': 'sum',
                                         'Average': 'sum',
                                         'SR' : 'max',
                                         'High Score': 'sum',
                                         'Innings by Bowling': 'sum',
                                         'Balls Bowled' : 'sum',
                                         'Maiden' : 'sum',
                                         'Runs Given' : 'sum',
                                         'Wkts' : 'sum',
                                         '4 W ' : 'sum',
                                         '5 W' : 'sum',
                                         'Catches' : 'sum',
                                         'Stumping' : 'sum',
                                         'Bol Average' : 'sum',
                                         'Bol SR' : 'max',
                                         'Econ' : 'min'}).reset_index()

player_stats.head()
```

```python
top_10_match = player_stats.nlargest(10, ['Match','Innings'])
top_10_match.head(1)
```

```python
top_10_match_pivot = top_10_match.pivot_table(index = 'Player', values = ['Match','Innings'])

top_10_match_pivot.plot.line(figsize = (10,6),marker = '*',grid = True)


tick = range(len(top_10_match_pivot))
plt.xticks(ticks = tick, labels = top_10_match_pivot.index, rotation = 45)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xlabel('Players')
plt.ylabel('No of Matches')
plt.title('Top 10 most Matches Played Player')

plt.tight_layout()
plt.show()
```

```python
top_10_runs = player_stats.nlargest(10, ['Runs','Balls Faced'])
top_10_runs.head(1)
```

```python
top_10_runs_pivot = top_10_runs.pivot_table(index = 'Player', values = ['Runs','Balls Faced'])

top_10_runs_pivot.plot.line(figsize = (10,6),marker = '*',grid = True)


tick = range(len(top_10_runs_pivot))
plt.xticks(ticks = tick, labels = top_10_runs_pivot.index, rotation = 45)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xlabel('Players')
plt.ylabel('Runs')
plt.title('Top 10 Run Geters')

plt.tight_layout()
plt.show()
```

```python
top_10_Ducks = player_stats.nlargest(10, ['Not Outs','Ducks'])
top_10_Ducks.head(1)

top_10_Ducks_pivot = top_10_Ducks.pivot_table(index = 'Player', values = ['Not Outs','Ducks'])

top_10_Ducks_pivot.plot.line(figsize = (10,6),marker = 'X',grid = True)


tick = range(len(top_10_Ducks_pivot))
plt.xticks(ticks = tick, labels = top_10_Ducks_pivot.index, rotation = 45)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xlabel('Players')
plt.ylabel('No of Ducks and Not Outs')
plt.title('Top 10 most Ducks and Not Outs by Player')

plt.tight_layout()
plt.show()
```

```python
top_10_4s = player_stats.nlargest(10, ['6s','4s'])
top_10_4s.head(1)

top_10_4s_pivot = top_10_4s.pivot_table(index = 'Player', values = ['6s','4s'])

top_10_4s_pivot.plot.line(figsize = (10,6),marker = 'X',grid = True)


tick = range(len(top_10_4s_pivot))
plt.xticks(ticks = tick, labels = top_10_4s_pivot.index, rotation = 45)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

plt.xlabel('Players')
plt.ylabel('No of Ducks and Not Outs')
plt.title('Top 10 most 6s and 4s by Player')

plt.tight_layout()
plt.show()
```
